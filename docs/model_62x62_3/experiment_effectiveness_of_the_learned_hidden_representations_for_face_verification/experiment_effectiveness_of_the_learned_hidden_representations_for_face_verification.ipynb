### Plot the result ###

[\[back to top\]](#Process)

````python
import matplotlib.pyplot as plt
import random
import matplotlib.markers as mks

def random_marker():
    markers = mks.MarkerStyle.markers
    num = len(markers.values())
    idx = random.randint(0, num - 1)
    return markers.values()[idx]


data_x = [264, 528, 1057, 2115, 4230, 8460]
data_y = [0.61, 0.60, 0.62, 0.63, 0.65, 0.66]

color = [random.random(), random.random(), random.random()]
marker = random_marker()
linewidth = 0.75
label = 'Joint Bayesian'

ok = False
## Some markers throw ValueError: Unrecognized marker style
while not ok:
    try:
        marker = random_marker()
        plt.plot(data_x, data_y, label = label, color = color, marker = marker, linewidth = linewidth)
        ok = True
    except:
        pass

legend_loc = 'lower right'
plt.legend(loc = legend_loc, ncol = 1) # ajust ncol to fit the space
plt.title('Test accuracy vs. People numer')
plt.xlabel('People numer')
plt.ylabel('Test accuracy')
plt.savefig('./test_accuracy_people_number.png')
````
